```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("../")
from pyACC.cosmology import Distances
from astropy import cosmology
```

```python
class HubbleFunction:
    def __init__(self, H0, OmegaM, OmegaL, OmegaR):
        self.H0 = H0
        self.OmegaM = OmegaM
        self.OmegaL = OmegaL
        self.OmegaR = OmegaR 

    def __call__(self, z):
        return self.H0 * np.sqrt(self.OmegaM * (1.0 + z)**3 + self.OmegaR * (1.0 + z)**4 + self.OmegaL)
        
```

```python
def hubble_function(z, H0, OmegaM, OmegaL, OmegaR):
    return H0 * np.sqrt(OmegaM * (1.0 + z)**3 + OmegaR * (1.0 + z)**4 + OmegaL)
```

```python
hf=HubbleFunction(70, 0.3, 0.7, 0.0)
d=Distances(hf)
for z in np.linspace(0, 100, 1000):
    plt.plot(z, d.comoving_distance(z), 'r.')
plt.xlabel("z")
plt.ylabel("Comoving distance [Mpc]")
plt.show()


```
